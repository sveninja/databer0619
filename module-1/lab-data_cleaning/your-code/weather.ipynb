## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd

df = pd.read_csv('../weather-raw.csv')
df.head()
```

```python
df.info()
```

```python
print(len(df))
null_cols = df.isnull().sum()
null_cols[null_cols > 0]
```

```python
drop_cols = null_cols[null_cols > 21].index
df = df.drop(drop_cols, axis = 1)
```

```python
print(drop_cols)
```

```python
df
```

```python
import datetime
```

```python
date_new = df['year'].astype(str) + '-' + df['month'].astype(str)

df.insert(3, 'Year-Month', date_new)
df
```

```python
list = []
for x in range(1, 32):
    if ("d" + str(x)) not in drop_cols:
        list.append("d" + str(x))
print(list)
```

```python
df_numeric = df[list]
df_numeric
```

```python
df_numeric['mean'] = df_numeric.mean(axis=1)
df_numeric.round(1)

# df['mean'] = df.mean(axis=1)
# df
```

```python
df_tmax = df_numeric[::2]
df_tmax.round(1)
```

```python
df_tmax_mean = df_tmax[['mean']]
df_tmax_mean = df_tmax_mean.reset_index(drop=True)
df_tmax_mean
```

```python
df_tmin = df_numeric[1::2]
df_tmin.round(1)
```

```python
df_tmin_mean = df_tmin[['mean']]
df_tmin_mean = df_tmin_mean.reset_index(drop=True)
df_tmin_mean
```

```python
df_half = df[::2]
df_half
```

```python
df_new = df_half[['id', 'Year-Month']]
df_new = df_new.reset_index(drop=True)
df_new
```

```python
# df_col = pd.concat([df1,df2], axis=1)
# df_col

df_new2 = pd.concat([df_new, df_tmax_mean, df_tmin_mean], axis=1)
df_new2[['id', 'Year-Month', 'mean', 'mean']]
df_new2.round(1)
```

```python
df_new2 = df_new2.set_axis(['ID', 'Year-Month', 'T max mean', 'T min mean'], axis=1, inplace=False)
df_new2.round(1)
```
